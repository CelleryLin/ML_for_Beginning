```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize=(8, 8))

n_dots = 120   # set number of dots
cirs = 100
random_amount = 0.5
angs = np.linspace(0, 2*np.pi*cirs, n_dots)  # angles to the dots
cx, cy = (0 , 0)  # center of circle
xs, ys = [], []    # for coordinates of points to plot
ra = 1.0          # radius of circle
data0 = []
for ang in angs:
    # compute (x,y) for each point
    x = cx + ra*np.cos(ang) + np.random.uniform(-random_amount, random_amount)
    y = cy + ra*np.sin(ang) + np.random.uniform(-random_amount, random_amount)
    xs.append(x)   # collect x
    ys.append(y)   # collect y
    data0.append([x,y])

plt.scatter(xs, ys)  # plot points 


angs = np.linspace(0, 2*np.pi*cirs, n_dots)  # angles to the dots
cx, cy = (0 , 0)  # center of circle
xs, ys = [], []    # for coordinates of points to plot
ra = 5.0          # radius of circle
cirs = 100
random_amount = 0.5
data1 = []
for ang in angs:
    # compute (x,y) for each point
    x = cx + ra*np.cos(ang) + np.random.uniform(-random_amount, random_amount)
    y = cy + ra*np.sin(ang) + np.random.uniform(-random_amount, random_amount)
    xs.append(x)   # collect x
    ys.append(y)   # collect y
    data1.append([x,y])

plt.scatter(xs, ys)  # plot points 


plt.show()
```

```python
# transform to polar coordinates
data0_polar = []
for i in range(len(data0)):
    x, y = data0[i]
    r = np.sqrt(x**2 + y**2)
    theta = np.arctan2(y, x)
    data0_polar.append([r, theta])

data1_polar = []
for i in range(len(data1)):
    x, y = data1[i]
    r = np.sqrt(x**2 + y**2)
    theta = np.arctan2(y, x)
    data1_polar.append([r, theta])

data0_polar = np.array(data0_polar).T
data1_polar = np.array(data1_polar).T

plt.scatter(data0_polar[0], data0_polar[1])  # plot points
plt.scatter(data1_polar[0], data1_polar[1])  # plot points
plt.xlim(0, 8)
plt.xlabel('r')
plt.ylabel('theta')
plt.show()
```
