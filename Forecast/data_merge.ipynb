```python
import pandas as pd
import numpy as np
from datetime import date, timedelta
import os
import sys
sys.path.append("./")

# import some tools
from _utils import create_dataset, data_preproc

csv_folder = './csvs/'
# all csv files in the folder
csv_files = [f for f in os.listdir(csv_folder) if os.path.isfile(os.path.join(csv_folder, f))]
csv_files = pd.DataFrame(sorted(csv_files), columns=['filename'])
print(csv_files)

datas = pd.read_csv('./data_tmp.csv')
mean_data = []
used_col = ['date', 'ObsTime', 'StnPres', 'SeaPres',
            'Temperature', 'Td dew point', 'RH', 'WS',
            'WD', 'WSGust', 'WDGust', 'Precp', 'PrecpHour',
            'UVI', 'Cloud Amount']

# merge data by mean every 24 hours
for i in range(0, len(datas), 24):
    mean_data.append({
        'date': datas['date'][i],
        # 'ObsTime': datas['ObsTime'][i],
        'StnPres': datas['StnPres'][i:i+24].mean(),
        'SeaPres': datas['SeaPres'][i:i+24].mean(),
        'Temperature': datas['Temperature'][i:i+24].mean(),
        'Td dew point': datas['Td dew point'][i:i+24].mean(),
        'RH': datas['RH'][i:i+24].mean(),
        'WS': datas['WS'][i:i+24].mean(),
        'WD': datas['WD'][i:i+24].mean(),
        'WSGust': datas['WSGust'][i:i+24].mean(),
        'WDGust': datas['WDGust'][i:i+24].mean(),
        'Precp': datas['Precp'][i:i+24].sum(),
        'PrecpHour': datas['PrecpHour'][i:i+24].sum(),
        'UVI': datas['UVI'][i:i+24].mean(),
        'Cloud Amount': datas['Cloud Amount'][i:i+24].mean()
    })

mean_data = pd.DataFrame(mean_data)
mean_data.to_csv('./mean_data.csv', index=False)
print(mean_data.head())
```
